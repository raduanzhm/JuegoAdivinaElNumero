```python
from ModuloMenuModos import MenuPrincipal
from ModuloMenuDificultad import Menudificultad
from ModuloJuego import JugarSolitario, Jugar2Jugadores
from EstadisticasJuego import guardar_partida, mostrar_resumen_y_grafico

def main():
    print('\n--- ADIVINA EL NUMERO ---')

    while True:
        opcion = MenuPrincipal()

        if opcion == 1:
            dificultad, intentos = Menudificultad()
            gano, usados, numero = JugarSolitario(intentos)
            jugador = input('Nombre del jugador: ').strip()
            guardar_partida(jugador, 'Solitario', dificultad, intentos, usados, gano, numero)
            print('Partida registrada.\n')

            if not gano:
                try:
                    intento = int(input("Ingresa tu ultimo intento para recibir una pista: "))
                    dar_pista(numero, intento)
                except ValueError:
                    print("Debes ingresar un numero valido.")

        elif opcion == 2:
            dificultad, intentos = Menudificultad()
            gano, usados, numero = Jugar2Jugadores(intentos)
            jugador = input('Nombre del jugador 2: ').strip()
            guardar_partida(jugador, '2 Jugadores', dificultad, intentos, usados, gano, numero)
            print('Partida registrada.\n')

        elif opcion == 3:
            print('\n--- ESTADISTICAS ---')
            mostrar_resumen_y_grafico()

        elif opcion == 4:
            print('\nGracias por jugar. Hasta pronto!')
            break

        else:
            print('Opcion no valida. Intenta de nuevo.')


if __name__ == '__main__':
    main()
```
